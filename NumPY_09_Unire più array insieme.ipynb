```python
import numpy as np

# Concatenate 1D e 2D

arr1 = np.array ([1,2,3])
arr2 = np.array ([5,6,7])

arr_con = np.concatenate((arr1,arr2))
print(arr_con)
```

```python
# 2D

arr1 = np.array ([[1,2] , [3,4]])
arr2 = np.array ([[5,6] , [7,8]])

arr_con = np.concatenate((arr1,arr2)) # axis = 0, cos√¨ concatena verticalmente
print(arr_con)
```

```python
arr1 = np.array ([[1,2] , [3,4]])
arr2 = np.array ([[5,6] , [7,8]])

arr_con = np.concatenate((arr1,arr2), axis = 1) # concatena orizzontalmente
print(arr_con)
```

```python
# Stack, hstack e vstack

arr1 = np.array ([1,2,3])
arr2 = np.array ([5,6,7])

arr_con = np.concatenate((arr1,arr2))
print(arr_con)

print("-" * 50)

arr_stacked = np.stack ((arr1,arr2), axis = 0)
print(arr_stacked)

print("-" * 50)

arr_stacked_ax1 = np.stack ((arr1,arr2), axis = 1)
print(arr_stacked_ax1)
```

```python
# "hstack" unisce in orizzontale

arr_stacked = np.hstack ((arr1,arr2))
print(arr_stacked)
```

```python
# "vstack" unisce in verticale

arr_stacked = np.vstack ((arr1,arr2))
print(arr_stacked)
```
