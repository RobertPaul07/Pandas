```python
import numpy as np

arr = np.array([1,2,3,4,5,6])
print(arr.dtype)
```

```python
arr1 = np.array(["a" , "b" , "c" , "d", "e"])
print(arr1.dtype)
```

```python
arr2 = np.array([False, True, False, True])
print(arr2.dtype)
```

```python
# Creare array con tipo "dtype e "astype"

arr_new = np.array([1,2,3,4,5,6] , dtype = "S")
print(arr_new)
```

```python
arr_2 = np.array(["1","2","3","4"], dtype = "i")
print(arr_2)
```

```python
# Astype

arr = np.array([1,2,3,4,5,6])
arr_1 = arr.astype("S")
print(arr_1)
```
