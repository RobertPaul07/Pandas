```python
import numpy as np

print(type(np.add)) # UFUNC , funzione universale
print(type(np.concatenate)) # funzione normale
```

```python
# Creare la propria UFUNC

arr = np.array([1,2,3,4,5,6])

def addCinque(x):
    return x + 5

# IMPORTANTE! Registrare funzione
addCinque = np.frompyfunc(addCinque, 1, 1) #nome funzione, n° argomenti input, n°output

print(addCinque(arr))
```

```python
lista_numeri = ([1,2,3,4,5,6])
print(addCinque(lista_numeri))

# funziona anche su lista numeri perchè ho REGISTRATO precedentemente la funz
# quindi in automatico converte la lista in un array
```

```python
# Altre funzioni universali

arr1 = np.array([10, 20, 30, 40, 50])
arr2 = np.array([5, 10, 15, 20, 25])

arr = np.add(arr1, arr2)
print(arr)
```

```python
arr = np.subtract(arr1, arr2)
print(arr)
```

```python
arr = np.multiply(arr1, arr2)
print(arr)
```

```python
arr = np.divide(arr1, arr2)
print(arr)
```

```python
arr = np.remainder(arr1, arr2)  # resto
print(arr)
```
