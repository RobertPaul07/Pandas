```python
import numpy as np

arr = np.array([1,2,3,4,5,6])

for x in arr:
    print(x)
```

```python
arr_2D = np.array([  [1,3,5,7] , [10,12,14,16]  ])

for x in arr_2D:
    for y in x:
        print(y)
```

```python
arr_3D = np.array([  [ [12,14,15],[18,19,20] ] , [ [1,2,3],[4,5,6] ]  ])

for x in arr_3D:
    for y in x:
        for z in y:
            print(z)
```

```python
# Scorciatoia con "nditer"

for x in np.nditer(arr_3D):
    print(x)
```
